```python
def issue_books(student_name,books):
  print(f"\n{student_name} borrowed:")
  for book in books:
    print("-",book)
def calculate_fine(days_late):
  fine_per_day=5
  return days_late*fine_per_day
def student_report(student_name,books,days_late):
  issue_books(student_name,books)
  fine=calculate_fine(days_late)
  print(f"Fine to be paid: ${fine}\n")

#example usuage
student_report("Rahul",["Python Basics","Data Science 101"],3)
student_report("Priya",["Machine Learning"],0)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
np.random.seed(42)
n=200
data=pd.DataFrame({
    "experience":np.random.randint(0,15,n),
    "education_years":np.random.randint(10,20,n),
    "age":np.random.randint(22,60,n)

})

data["salary"]=(
    30000+
    data["experience"]*4000+
    data["experience"]*2000+
    data["age"]*500+
         np.random.randn(n)*10000
)

data.head()
```

```python
print(data.shape)
print(data.info())
print(data.describe())
```

```python
corr=data.corr()
plt.imshow(corr)
plt.xticks(range(len(corr.columns)),corr.columns,rotation=45)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.show()
```

```python
features=["experience","education_years","age"]

for col in features:
  plt.figure()
  plt.scatter(data[col],data["salary"])
  plt.xlabel(col)
  plt.ylabel("salary")
  plt.show()
```

```python
x=data[["experience","education_years","age"]]
y=data["salary"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)

print(mse,rmse,r2)
```

```python
new_dta=pd.DataFrame({
    "experience":[5],
    "education_years":[16],
    "age":[30]
})

print(model.predict(new_dta))
```
